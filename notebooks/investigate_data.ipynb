```python
import pandas as pd
import numpy as np
import utils
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/WDBC.csv")
columns = utils.get_column_categories()
centralized_df = utils.centralize(df, columns['numerical'])
normalized_df  = utils.normalize(df, columns['numerical'])
```

```python
fig, axs = plt.subplots(2,5)
fig.set_size_inches(15,7)
axi = [0,0]
for col in columns['worst']:
    sns.histplot(data = df, x = col, hue = 'diagnosis', ax = axs[axi[0],axi[1]] , stat = 'probability')
    axi[1] = axi[1]+1
    if axi[1] == 5:
        axi[1] = 0
        axi[0] = 1
```

```python
fig, axs = plt.subplots(2,5)
fig.set_size_inches(15,7)
axi = [0,0]
for col in columns['mean']:
    sns.histplot(data = normalized_df, x = col, ax = axs[axi[0],axi[1]] , stat = 'probability')
    axi[1] = axi[1]+1
    if axi[1] == 5:
        axi[1] = 0
        axi[0] = 1
```
